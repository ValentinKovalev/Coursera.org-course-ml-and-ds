```python
import pandas as pd
```

```python
data = open('checkins.dat', 'r')
result = list()
with open('checkins.csv', "w") as f:
    for line in data:
        #print(line)
        if 'id' not in line:
            last = line.split()[-1]
            result_line = ''.join(line.split()[:-1]) + "T" + last + '\n'
        elif 'id' in line:
            result_line = ''.join(line.split()) + '\n'
        f.write(result_line)
```

```python
df = pd.read_csv('checkins.csv', sep='|')
df = df.drop(['created_at'], axis = 1)
df = df.loc[:, 'latitude':'longitude']
print(df.shape)
df = df.dropna()
print(df.shape)
df.head(5)
df = df.iloc[0:100000]
print(df.shape)
```

```python
from sklearn.cluster import MeanShift

clustering_alg = MeanShift(bandwidth=0.1).fit(df)
```

```python
labels = clustering_alg.labels_
cluster_centers = clustering_alg.cluster_centers_
```

```python
amount_of_cluster_centers = {}
for label in labels:
    if label in amount_of_cluster_centers:
        amount_of_cluster_centers[label] +=1
    else:
        amount_of_cluster_centers[label] = 1

good_clusters_labels = {}
for key, value in amount_of_cluster_centers.items():
    if value > 15:
        good_clusters_labels[key] = value
good_clusters_labels
```

```python
good_clusters_labels = good_clusters_labels.keys()
```

```python
result_cluster_centers = []
print(labels)
for label, cluster_center in zip(labels, cluster_centers):
    if label in good_clusters_labels:
        result_cluster_centers.append(cluster_center)
result_cluster_centers
```
