```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
import scipy.stats
%matplotlib inline
import math
```

```python
a = 1.99
mean, var, skew, kurt = gamma.stats(a, moments='mvsk')

E = a
D = a ** 2
```

```python
x = np.linspace(gamma.ppf(0.01, a),
                 gamma.ppf(0.99, a), 100)
plt.plot(x, gamma.pdf(x, a),
        'r-', lw=5, alpha=0.6, label='gamma pdf')
plt.show()
```

```python
#генерируем выборку объёма 1000
a = 1.99 #параметр для гамма-функции
gamma_func = gamma(a) #ининциализируем гамма-функцию с заданным параметром
sample_len = 1000 #инициализируем длину выборки
sample = gamma_func.rvs(sample_len) #выборка гамма-функции из sample_len значений
x = np.linspace(1, 10, sample_len) #массив точек для получения значений плотности распределения
pdf = gamma_func.pdf(x) #ищем значение плотности распределения от выборки
#строим гистрограмму выборки и теоретическую функцию распределения
plt.plot(x, pdf, label = 'theoretical PDF')
plt.hist(sample, bins = 50, range = (1,10), density = True, label = 'practic pdf hist')
plt.ylabel('$f(x)$/number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```

```python
a = 1.99 #параметр для гамма-распределения
gamma_func = gamma(a) #инициализация функции гамма-распределения
n = [5, 10, 50] #объёмы выборки из условия
for element in n:
    sample_n = np.array([])
    i = 1
    i_max = 1000
    while i <= i_max:
        sample_n_i = gamma_func.rvs(element)
        mean_n_i = sum(sample_n_i) / element
        sample_n = np.append(sample_n, mean_n_i)
        i += 1
    E = a
    D = a ** 2
    sample_n_all = np.array([])
    norm = scipy.stats.norm(E, math.sqrt(D / element))
    pdf = norm.pdf(x[:1000])
    sample_n_all = np.append(sample_n_all, sample_n)
    plt.hist(sample_n, bins = 100, color = 'blue', range=(1, 10), normed=True, alpha = 0.35, label = 'Gamma hist for n='+str(element))
    plt.ylabel('number of samples')
    plt.xlabel('$x$')
    plt.legend(loc='best')
    plt.plot(x[:1000], pdf)
    plt.show()
```

Для выполнения задания была выбрана функция гамма-распределения. 

Были выполнены условия задачи: построены для четырех разных n (5, 10, 50) по 1000 случайных выборок, измерены их выборочные средние и составлены соостветствующие массивы, по массивам выборочных средних построены гистограммы распределения, исходя из теоретических данных определены соответствующие каждому n параметры распределений для апроксимации, и по полученным параметрам построены функции нормального распределения для каждого n поверх соответствующих гистограмм.

По полученным графикам наблюдается увелечение точности апроксимации выборочного среднего значения непрерывной случайной величины описываемой гамма-распределением, хотя и в итоговый результат внесены искажения из-за особенностей функции распределения, главным из которых является граница значений, что приводит к ограничению искомой апроксимирующей функции. Также заметна скошенность проявляющаяся в несимметричности распределения относительно центра. 
Из-за перечисленных недостатков функция сходится медленно, и более качественный результат наблюдается при больших n.
